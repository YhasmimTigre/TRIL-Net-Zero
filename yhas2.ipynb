```python
import numpy as np, pickle as pkl, matplotlib.pyplot as plt

def picasso(n,file_out):
    img  = np.cos(np.random.rand(n,n))

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if i % 2 == 0 and j % 4 == 0:
                img[i,j] = 0        


    # salva arquivo pickle
    file_out = open(file_out,'wb')
    pkl.dump(img,file_out)

    return img

n = 50
img = picasso(n,'matrix.pkl')

# visualizao 
plt.figure(figsize=(10,10))
plt.matshow(img)            
```

```python
# recarregar
file_in = open('matrix.pkl','rb')
img = pkl.load(file_in)
```
