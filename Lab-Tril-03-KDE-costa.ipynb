# 2 - Aplicando o KDE

```python
from numpy import array, linspace
from sklearn.neighbors import KernelDensity
from matplotlib.pyplot import plot
```

```python
min_i = 0
max_i1 = J1_costa["Jc1"].shape[0]

aux1 = max_i1 - min_i
print("Numero de registros avaliados: ", aux1)
X1 = J1_costa["Jc1"].values[::].reshape(-1, 1)
#len

kde1 = KernelDensity(kernel='gaussian', bandwidth= 0.01).fit(X1) #aplica o KDE
s1 = np.linspace(0, 1.0, aux1) #Retorna nmeros uniformimente espaados em um intervalo especificado.
e1 = kde1.score_samples(s1.reshape(-1,1)) #Calcule a probabilidade logartmica de cada amostra sob o modelo

plt.figure(figsize=(20, 12))
plt.title("Kernels | J1 Costa", fontsize=22)
plt.ylabel("Densidade", fontsize=16)
plt.xlabel("Valores assumidos por J1 Costa", fontsize=16)
plt.grid(True)

plt.plot(s1, e1)
```

# 3 - Encontrando vales

```python
from scipy.signal import argrelmin
```

```python
#Determinando o KDE do J1 costa

min_i = 0
max_i1 = J1_costa["Jc1"].shape[0]
aux1 = max_i1 - min_i
J1 = J1_costa["Jc1"].values[::].reshape(-1, 1)
kde_J1 = KernelDensity(kernel='gaussian', bandwidth=0.01).fit(J1)
s1 = np.linspace(0, 1.0, aux1)
e1 = kde_J1.score_samples(s1.reshape(-1,1))
x1 = e1
valleys1 = argrelmin(x1)[0]
ord1 = np.argsort(np.abs(np.diff(x1[valleys1])))
ordValleys_J1 = np.flip(ord1)+1
```

```python
x1 = e1
valleys1 = argrelmin(x1)[0]

plt.figure(figsize=(20, 12))
plt.title("Vales dos Kernels | J1 Costa", fontsize=22)
plt.ylabel("Densidade", fontsize=16)
plt.xlabel("Posio assumida por J1 Costa", fontsize=16)
plt.grid(True)
plt.plot(x1)
plt.plot(valleys1, x1[valleys1], 'o') #Aqui determinamos os vales
plt.plot(valleys1[ordValleys_J1[0:4]], x1[valleys1[ordValleys_J1[0:4]]], '*', c='green') #aqui determinamos onde os pontos de diviso devem aparecer
plt.show()

print(valleys1)
```
